```python
from transformers import MarianMTModel, MarianTokenizer

# Function to translate text
def translate(text, src_lang, tgt_lang):
    model_name = f'Helsinki-NLP/opus-mt-{src_lang}-{tgt_lang}'  # Model for translation
    model = MarianMTModel.from_pretrained(model_name)
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    
    # Tokenize and encode the input text
    inputs = tokenizer.encode(text, return_tensors="pt", padding=True)
    
    # Generate the translation
    translated = model.generate(inputs, max_length=100)
    
    # Decode the translated text
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    
    return translated_text

# Example Translations
```

```python
input_text = "My name is Bilal?"
input_textD ="Wie war dein Tag heute?"
input_textF="Quel est ton plat préféré ?"
# English to German
translated_text_de = translate(input_text, 'en', 'de')
print(f"English to German: {translated_text_de}")

# English to French
translated_text_fr = translate(input_text, 'en', 'fr')
print(f"English to French: {translated_text_fr}")

# German to English
translated_text_en_from_de = translate(input_textD, 'de', 'en')
print(f"German to English: {translated_text_en_from_de}")

# French to English
translated_text_en_from_fr = translate(input_textF, 'fr', 'en')
print(f"French to English: {translated_text_en_from_fr}")
```
